**Dataset Overview and Key Relationships**

An initial examination of the provided datasets reveals the following structures and interrelationships, crucial for subsequent analysis:

- The **`purchases`** table records individual procurement transactions, detailing the date of purchase, the specific brands acquired from vendors, the corresponding monetary value (`Dollars`), and the volume (`Quantity`).
- The **`purchase_prices`** table serves as a lookup for product pricing, providing both the retail price (`Price`) and the cost to the vendor (`PurchasePrice`) at the brand level. The uniqueness of this table is established by the combination of `VendorNumber` and `Brand`.
- The **`vendor_invoice`** table represents an aggregated view of procurement activities, summarizing total quantity and dollar amounts from associated purchases. It also includes `Freight` costs and is uniquely identified by the combination of `VendorNumber` and `PONumber`.
- The **`sales`** table captures consumer sales transactions, including the brands sold, the volume (`SalesQuantity`), the selling price (`SalesPrice`), and the generated revenue (`SalesDollars`). This table is fundamental for analyzing sales performance and trends.
